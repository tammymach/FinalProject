Step 2: Removing Bad Data

When looking through the dataset, we notice that four of the entries in the "Mass" column are above 10,000g, and are entered using "XX,XXX" format. We have discovered Python runs into issues trying to turn these strings into floats, and so we are removing the four entries that make use of this formatting.

```python
#Looks for entries in the "mass" column entered with a comma, records how many there are and where they are (what row)
Word = "a"
Counters = 0
WordPos = []
for x in range(len(BirdArray)):
    Word = BirdArray[x, 1]
    if Word.find(",") > 0:
        Counters += 1
        WordPos.append(x)

print(Counters)
print(WordPos)
```

```python
#Removes the rows that were found using the above method, hard-coded with their locations

BirdMerge2 = BirdMerge.drop(BirdMerge.index[[161, 162, 220, 394]]) 
# http://stackoverflow.com/questions/14661701/how-to-drop-a-list-of-rows-from-pandas-dataframe
BirdMerge2
```

Step 3: Initial Plots

We have plotted here masses of the birds as a variable of percentage of foraging strategy employed. There are four foraging strategies (ground, air, forest, and water), and thus four graphs.

```python
GroundAmt = []
WhichBirdGround = []

#Goes through and finds all rows in which the amount of foraging in the Ground column (the -15 one) is greater than 0,
#indicating at least some foraging on the ground
for x in range (len(BirdArray)):
    if BirdArray[x, -15] > 0:
        GroundAmt.append(BirdArray[x, -15])
        WhichBirdGround.append(x)

#Plots the masses vs. the % of foraging done on the ground        
for GroundBird in range(len(GroundAmt)):
    plt.scatter(GroundAmt[GroundBird], BirdArray[WhichBirdGround[GroundBird], 1])


#Fills arrays with the masses and amount of foraging done on the ground
GroundBirdMass = np.zeros(len(GroundAmt))
GroundBirdAmt = np.zeros(len(GroundAmt))
for x in range(len(GroundAmt)):
    GroundBirdMass[x] = BirdArray[WhichBirdGround[x], 1]
    GroundBirdAmt[x] = GroundAmt[x]
    
plt.xlabel("% Foraging on Ground")
plt.ylabel("Mass (g)")
plt.title("Masses of Ground-Foraging Birds")
plt.xlim(0, 110)
plt.ylim(0, max(GroundBirdMass)+100)
plt.savefig("GroundForaging.jpg")
```

```python
#Does the same as the ground coding, but for the air column and thus air-foragers
AirAmt = []
WhichBirdAir = []
for x in range (len(BirdArray)):
    if BirdArray[x, -11] > 0:
        AirAmt.append(BirdArray[x, -11])
        WhichBirdAir.append(x)
        
for AirBird in range(len(AirAmt)):
    plt.scatter(AirAmt[AirBird], BirdArray[WhichBirdAir[AirBird], 1])

AirBirdMass = np.zeros(len(AirAmt))
AirBirdAmt = np.zeros(len(AirAmt))
for x in range(len(AirAmt)):
    AirBirdMass[x] = BirdArray[WhichBirdAir[x], 1]
    AirBirdAmt[x] =AirAmt[x]
    
plt.xlabel("% Foraging in Air")
plt.ylabel("Mass (g)")
plt.title("Masses of Air-Foraging Birds")
plt.xlim(0, 110)
plt.ylim(0, max(AirBirdMass)+100)
plt.savefig("AirForaging.jpg")
```

```python
#Does the same process of finding the forest-foraging birds and plotting their masses to the amount of forest-foraging,
#with the added caveat that it combines three columns that are all specific zones of forest (understory, midhigh, canopy)
ForestAmt = []
WhichBirdForest = []
for x in range (len(BirdArray)):
    if (BirdArray[x, -14] + BirdArray[x, -13] + BirdArray[x, -12]) > 0:
        ForestAmt.append((BirdArray[x, -14] + BirdArray[x, -13] + BirdArray[x, -12]))
        WhichBirdForest.append(x)
        
for ForestBird in range(len(ForestAmt)):
    plt.scatter(ForestAmt[ForestBird], BirdArray[WhichBirdForest[ForestBird], 1])
   
ForestBirdMass = np.zeros(len(ForestAmt))
ForestBirdAmt = np.zeros(len(ForestAmt))
for x in range(len(ForestAmt)):
    ForestBirdMass[x] = BirdArray[WhichBirdForest[x], 1]
    ForestBirdAmt[x] = ForestAmt[x]
    
plt.xlabel("% Foraging in Forest")
plt.ylabel("Mass (g)")
plt.title("Masses of Forest-Foraging Birds")
plt.xlim(0, 110)
plt.ylim(0, max(ForestBirdMass)+100)
plt.savefig("ForestForaging.jpg")
```

```python
#Again, finds and plots the masses and % foraging for birds foraging at least partially in water. Also makes use of two
#columns that are both under the umbrella of "water": below surf and around surf
WaterAmt = []
WhichBirdWater = []
for x in range (len(BirdArray)):
    if (BirdArray[x, -15] + BirdArray[x, -16]) > 0:
        WaterAmt.append((BirdArray[x, -15] + BirdArray[x, -16]))
        WhichBirdWater.append(x)
        
for WaterBird in range(len(WaterAmt)):
    plt.scatter(WaterAmt[WaterBird], BirdArray[WhichBirdWater[WaterBird], 1])
   
WaterBirdMass = np.zeros(len(WaterAmt))
WaterBirdAmt = np.zeros(len(WaterAmt))
for x in range(len(WaterAmt)):
    WaterBirdMass[x] = BirdArray[WhichBirdWater[x], 1]
    WaterBirdAmt[x] = WaterAmt[x]
    
plt.xlabel("% Foraging in Water")
plt.ylabel("Mass (g)")
plt.title("Masses of Water-Foraging Birds")
plt.xlim(0, 110)
plt.ylim(0, max(WaterBirdMass)+100)
plt.savefig("WaterForaging.jpg")
```

Step 4: Linear Regression Models and Pearson Test

Here we are fitting linear regression models to our data to see what sort of trend there may be with the percentage of foraging strategy and the body masses of the birds. We are also running a Pearson correlation test to determine the amount of correlation in the data (how much variation in Y is explained by variation in X), as well as if this is a statistically significant correlation.

```python
from sklearn.linear_model import LinearRegression
from scipy.stats.stats import pearsonr

modelGround = LinearRegression(fit_intercept=True)

X = GroundBirdAmt[:, np.newaxis]
y = GroundBirdMass

modelGround.fit(X, y)

plt.plot(X, y, '.')

xx = np.linspace(1, 100)
yy = modelGround.predict(xx[:, np.newaxis])
plt.plot(xx, yy, 'g')

plt.title("Linear Regression Model of Ground-Foraging Birds")
plt.xlabel("% Ground Foraging")
plt.ylabel("Body Mass (g)")
plt.xlim(0, 110)
plt.text(10, 4500, "slope: " + str(modelGround.coef_))
plt.text(10, 4200, "Y-intercept: " + str(modelGround.intercept_))


GroundPear = pearsonr(GroundBirdAmt, GroundBirdMass)
plt.text(10, 3600, "Pearson Correlation: " + str(GroundPear[0]))
plt.text(10, 3300, "p-value: " + str(GroundPear[1]))

plt.savefig("GroundRegression.jpg")
```

```python
modelAir = LinearRegression(fit_intercept=True)

X = AirBirdAmt[:, np.newaxis]
y = AirBirdMass

modelAir.fit(X, y)

plt.plot(X, y, '.')

xx = np.linspace(1, 100)
yy = modelAir.predict(xx[:, np.newaxis])
plt.plot(xx, yy, 'g')

plt.title("Linear Regression Model of Air-Foraging Birds")
plt.xlabel("% Air Foraging")
plt.ylabel("Body Mass (g)")
plt.xlim(0, 110)
plt.text(40, 1400, "slope: " + str(modelAir.coef_))
plt.text(40, 1300, "Y-intercept: " + str(modelAir.intercept_))


AirPear = pearsonr(AirBirdAmt, AirBirdMass)
plt.text(40, 1000, "Pearson Correlation: " + str(AirPear[0]))
plt.text(40, 900, "p-value: " + str(AirPear[1]))

plt.savefig("AirRegression.jpg")
```

```python
modelForest = LinearRegression(fit_intercept=True)

X = ForestBirdAmt[:, np.newaxis]
y = ForestBirdMass

modelForest.fit(X, y)

plt.plot(X, y, '.')

xxf = np.linspace(1, 100)
yyf = modelForest.predict(xxf[:, np.newaxis])
plt.plot(xxf, yyf, 'g')

plt.title("Linear Regression Model of Forest-Foraging Birds")
plt.xlabel("% Forest Foraging")
plt.ylabel("Body Mass (g)")
plt.xlim(0, 110)
plt.text(10, 2300, "slope: " + str(modelForest.coef_))
plt.text(10, 2100, "Y-intercept: " + str(modelForest.intercept_))


ForestPear = pearsonr(ForestBirdAmt, ForestBirdMass)
plt.text(40, 1800, "Pearson Correlation: " + str(ForestPear[0]))
plt.text(40, 1600, "p-value: " + str(ForestPear[1]))

plt.savefig("ForestRegression.jpg")
```

Step 6: Comparing Predictions to Actual Data

Since we do have the BMR data, we are going to see how well our predicted BMR does when comparing to the actual BMR. We will be using means here. Predicting BMR will be done by first using the regression models to predict the mass for a bird that does x% foraging. Those predicted masses will then be put into an equation obtained from McNab 2009, one of the datasets for this project. This will give a value of predicted BMR for each of the % of foraging strategies, which can then be compared to actual mean BMR at that %.

```python
#Determining which percentages are present in the data for water-foraging birds
set(WaterBirdAmt)
```

```python
#Creating an array based on which percentages are present
WaterBirdSet = [10.0, 20.0, 25.0, 30.0, 33.0, 40.0, 50.0, 60.0, 70.0, 80.0, 90.0, 100.0]
```

```python
MeanBMRs = []
#This function takes an argument, "Amt", which is used to indicate the amount (or %) of foraging done by the bird
#It will calculate the mean BMR of all the birds doing that particular amount of foraging
def MeanBMRPercent(Amt):
    Percent = []
    for x in range(len(WaterAmt)):
        if WaterAmt[x] == Amt:
            Percent.append(x)
    BMRs = []
    for y in range(len(Percent)):
        BMRs.append(BirdArray[WhichBirdWater[Percent[y]], 2])
            
    return np.mean(BMRs)
```

```python
#Creating the array containing the mean BMR for each % found in the set
for x in range(len(WaterBirdSet)):
    MeanBMRs.append(MeanBMRPercent(WaterBirdSet[x]))
```

```python
#Predicts the mass of the birds, based off the linear regression model
PredictBirdMass = []
for t in range(len(WaterBirdSet)):
    PredictBirdMass.append(5.80*WaterBirdSet[t]-19.75)
```

```python
#Predicts BMR based off the predicted mass and the equation found in McNab 2009 for calculating BMR: 0.145m^.652, where
#m is the mass in grams
PredictBMR = []
for y in range(len(PredictBirdMass)):
    PredictBMR.append(0.145*PredictBirdMass[y]**(652/1000))
```

```python
#Plots the actual mean BMRs and the predicted BMRs
plt.scatter(WaterBirdSet, MeanBMRs, color = 'orange', label = "Actual Data")
plt.scatter(WaterBirdSet, PredictBMR, color = 'blue', label = "Predicted Data")
plt.legend(loc = "upper left")
plt.title("Predicted vs. Actual Mean BMR for Water-Foraging Birds")
plt.xlabel("% Foraging in Water")
plt.ylabel("BMR (kJ/h)")
plt.xlim(0, 110)
plt.savefig("WaterComparison.jpg")
```

Step 7: Changing the Models for Water-Foraging Birds to Account for Climate

We now have a comparison of the predicted BMR for both water- and forest-foraging birds. However, this is taking into account just the one variable. What happens if we account for another? As there were few data points that varied from the norm in most graphs, we have chosen to look at how climate may affect the estimates of BMR in water-foraging birds. 

```python
#This entire portion is finding the masses and the % of foraging for the birds specifically for each of the four
#climates: polar, temperate, tropic, and temperate/tropic
PolarWaterBirdMass = []
for r in range(len(PWBirds)):
    PolarWaterBirdMass.append(float(BirdArray[WhichBirdWater[PWBirds[r]], 1]))

PWBirdWaterAmt = np.zeros(len(PWBirds))
for t in range(len(PWBirds)):
    PWBirdWaterAmt[t] = (WaterAmt[PWBirds[t]])
    
TempWaterBirdMass = []
for r in range(len(TWBirds)):
    TempWaterBirdMass.append(float(BirdArray[WhichBirdWater[TWBirds[r]], 1]))

TWBirdWaterAmt = np.zeros(len(TWBirds))
for t in range(len(TWBirds)):
    TWBirdWaterAmt[t] = (WaterAmt[TWBirds[t]])
    
TropicWaterBirdMass = []
for r in range(len(TrWBirds)):
    TropicWaterBirdMass.append(float(BirdArray[WhichBirdWater[TrWBirds[r]], 1]))

TrWBirdWaterAmt = np.zeros(len(TrWBirds))
for t in range(len(TrWBirds)):
    TrWBirdWaterAmt[t] = (WaterAmt[TrWBirds[t]])
    
TempTropicWaterBirdMass = []
for r in range(len(TTrWBirds)):
    TempTropicWaterBirdMass.append(float(BirdArray[WhichBirdWater[TTrWBirds[r]], 1]))

TTrWBirdWaterAmt = np.zeros(len(TTrWBirds))
for t in range(len(TTrWBirds)):
    TTrWBirdWaterAmt[t] = (WaterAmt[TTrWBirds[t]])
```

```python
#Fitting linear regression models that take into account the climates
modelPWater = LinearRegression(fit_intercept=True)
modelTWater = LinearRegression(fit_intercept=True)
modelTrWater = LinearRegression(fit_intercept=True)
modelTTrWater = LinearRegression(fit_intercept=True)

X = PWBirdWaterAmt[:, np.newaxis]
y = PolarWaterBirdMass

A = TWBirdWaterAmt[:, np.newaxis]
b = TempWaterBirdMass

L = TrWBirdWaterAmt[:, np.newaxis]
m = TropicWaterBirdMass

R = TTrWBirdWaterAmt[:, np.newaxis]
s = TempTropicWaterBirdMass

modelPWater.fit(X, y)
modelTWater.fit(A, b)
modelTrWater.fit(L, m)
modelTTrWater.fit(R, s)

plt.plot(X, y, '.', color = 'black')
plt.plot(A, b, '.', color = 'red')
plt.plot(L, m, '.', color = 'green')
plt.plot(R, s, '.', color = 'cyan')

xxwp = np.linspace(1, 100)
yywp = modelPWater.predict(xxwp[:, np.newaxis])
xxwt = np.linspace(1, 100)
yywt = modelTWater.predict(xxwt[:, np.newaxis])
xxwtr = np.linspace(1, 100)
yywtr = modelTrWater.predict(xxwtr[:, np.newaxis])
xxwttr = np.linspace(1, 100)
yywttr = modelTTrWater.predict(xxwttr[:, np.newaxis])

plt.plot(xxwp, yywp, color = 'black')
plt.plot(xxwt, yywt, color = 'red')
plt.plot(xxwtr, yywtr, color = 'green')
plt.plot(xxwttr, yywttr, color = 'cyan')
plt.xlim(0, 110)
plt.title("Linear Regressions for Water-Foraging Birds by Climate")
plt.xlabel("% Foraging in Water")
plt.ylabel("Body mass (g)")
plt.legend(handles = [black_patch, red_patch, green_patch, cyan_patch], loc = "upper left")
plt.savefig("WaterRegressionsClimate.jpg")
```
